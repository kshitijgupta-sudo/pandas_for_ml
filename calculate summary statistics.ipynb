```python
import pandas as pd
```

```python
titanic = pd.read_csv("data/titanic.csv")
```

```python
titanic.head()
```

```python
titanic["Age"].mean()
```

```python
titanic[["Age","Fare"]].median()
```

```python
#aggregating statistic can be calculated for multiple columns at the same time
```

```python
titanic.describe()
```

```python
titanic.agg(
    {
        "Age": ["min", "max", "median", "skew"],
        "Fare": ["min", "max", "median", "mean"],
    }
)
```

```python
titanic.groupby("Sex").mean(numeric_only=True)
```

```python
#Count number of records by category
```

```python
#What is the number of passengers in each of the cabin classes?

titanic["Pclass"].value_counts()
```

```python
#well this is a good start but we can also use groupby to get the same result(long way)
titanic.groupby("Pclass").count()["PassengerId"]
```

```python
#Both size and count can be used in combination with groupby.
#  Whereas size includes NaN values and just provides the number of rows (size of the table), 
# count excludes the missing values
# . In the value_counts method, use the dropna argument to include or exclude the NaN values
```
