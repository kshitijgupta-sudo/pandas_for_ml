```python
import pandas as pd
```

```python
air_quality_no2 = pd.read_csv("data/air_quality_no2_long.csv",
                              parse_dates=True)
```

```python
air_quality_no2.head()
```

```python
air_quality_no2 = air_quality_no2[["date.utc","location","parameter","value"]]
```

```python
air_quality_no2.head()
```

```python
air_quality_pm25 = pd.read_csv("data/air_quality_pm25_long.csv")
```

```python
air_quality_pm25.head()
```

```python
air_quality_pm25 = air_quality_pm25[["date.utc","location","parameter","value"]]
```

```python
air_quality_pm25.head()
```

```python
#combine the measurements of 𝑁⁢𝑂2 and 𝑃⁢𝑀25 ,combine the measurements of 𝑁⁢𝑂2 and p⁢𝑀25
```

```python
air_quality = pd.concat([air_quality_pm25,air_quality_no2],axis = 0)
```

```python
air_quality.head()
```

```python
#By default concatenation is along axis 0, so the resulting table combines the rows of the input tables. 
```

```python
print("Shape of the air_quality_pm25 table:",air_quality_pm25.shape)
```

```python
print("Shape of the air_quality_no2 table:",air_quality_no2.shape)
```

```python
print('Shape of the resulting ``air_quality`` table: ', air_quality.shape)
```

```python
#Most operations like concatenation or summary statistics are by default across rows (axis 0)
```

```python
air_quality = air_quality.sort_values("date.utc")
air_quality.head()
```

```python
#The concatfunction provides a convenient solution with the keys argument, adding an additional (hierarchical) row index
```

```python
air_quality = pd.concat([air_quality_pm25, air_quality_no2],keys= ["PM25","no2"])
```

```python
##Join tables using a common identifier
```

```python
#Add the station coordinates, provided by the stations metadata table, to the corresponding rows in the measurements table.
```

```python
station_coord = pd.read_csv("data/air_quality_stations.csv")
```

```python
station_coord.head()
```

```python
air_quality.head()
```

```python
air_quality = pd.merge(air_quality , station_coord , how = "left", on = "location")
```

```python
air_quality.head()
```

```python
#Using the merge() function, for each of the rows in the air_quality table, the corresponding coordinates are added from the 
#air_quality_stations_coord table. Both tables have the column location in common which is used as a key to combine the information.
#By choosing the left join, only the locations available in the air_quality (left) table, i.e.
#FR04014, BETR801 and London Westminster, end up in the resulting table. 
```

```python
air_quality_parameters = pd.read_csv("data/air_quality_parameters.csv")
```

```python
air_quality_parameters.head()
```

```python
air_quality = pd.merge(air_quality,air_quality_parameters,how= "left",left_on= "parameter",right_on="id")
```
