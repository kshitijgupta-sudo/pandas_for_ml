```python
import pandas as pd 
#abe chutiye matplotlib is a base 
#matplotlib is the giant parent library, but it doesn't actually contain the subplots tool directly on its surface.
#  subplots lives inside a specific toolset within that library called pyplot. 
# By importing matplotlib.pyplot as plt, you give yourself access to everything you need for plotting.
import matplotlib.pyplot as plt
```
